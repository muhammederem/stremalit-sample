```python
from sklearn.preprocessing import OneHotEncoder


def preprocess(data):
    target = data['class']
    data = data.drop('class', axis=1)
    #target
    target = encoder.fit_transform(target)
    #set column as y column
    target = pd.DataFrame(target, columns=['class'])

    max_val = max_values(data)
    min_val = min_values(data)

    gender = OneHotEncoder().fit_transform(data['gender'])

    data['age'] = scaler.fit_transform(data['age'].values.reshape(-1,1))
    data['height_cm'] = scaler.fit_transform(data['height_cm'].values.reshape(-1,1))
    data['weight_kg'] = scaler.fit_transform(data['weight_kg'].values.reshape(-1,1))
    data['body fat_%'] = scaler.fit_transform(data['body fat_%'].values.reshape(-1,1))
    data['diastolic'] = scaler.fit_transform(data['diastolic'].values.reshape(-1,1))
    data['gripForce'] = scaler.fit_transform(data['gripForce'].values.reshape(-1,1))
    data['sit and bend forward_cm'] = scaler.fit_transform(data['sit and bend forward_cm'].values.reshape(-1,1))
    data['sit-ups counts'] = scaler.fit_transform(data['sit-ups counts'].values.reshape(-1,1))
    data['broad jump_cm'] = scaler.fit_transform(data['broad jump_cm'].values.reshape(-1,1))

    return data, target, max_val, min_val

data, target, max_values, min_values = preprocess(data)
```

```python
def remove_correlated(data):
    correlation = data.corr()
    for i in range(len(correlation.columns)):
        for j in range(i):
            if abs(correlation.iloc[i, j]) > 0.85:
                colname = correlation.columns[i]
                data.drop(colname, axis=1, inplace=True)
    return data
```

```python
#find outliers and remove them
def find_outliers(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3-q1
    lower_bound = q1 -(1.5 * iqr)
    upper_bound = q3 +(1.5 * iqr)

    #fill outliers with median
    for column in data.columns:
        data[column] = np.where(data[column] > upper_bound[column], data[column].median(), data[column])
        data[column] = np.where(data[column] < lower_bound[column], data[column].median(), data[column])

    return data

data = remove_correlated(data)
data = find_outliers(data)

data.info()
```

```python
#most important features
def most_important_features(data, target,threshold):
    from sklearn.ensemble import ExtraTreesClassifier
    import matplotlib.pyplot as plt
    model = ExtraTreesClassifier()
    model.fit(data, target)
    for feature in zip(data.columns, model.feature_importances_):
        if feature[1] > threshold:
            data.drop(feature[0], axis=1, inplace=True)
    return data
```
